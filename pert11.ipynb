```python
import pandas as pd

df = pd.read_csv("DataSampah.csv")
df
```

```python
tahun_pilih = 2016
total_sampah = 0

for i, row in df.iterrows():
    if row['tahun'] == tahun_pilih:
        total_sampah += row['jumlah_sampah']

print(f"Total produksi sampah Jawa Barat tahun {tahun_pilih}: {total_sampah:.2f} ton")
```

```python
total_per_tahun = {}

for i, row in df.iterrows():
    tahun = row['tahun']
    jumlah = row['jumlah_sampah']
    
    if tahun not in total_per_tahun:
        total_per_tahun[tahun] = 0
    
    total_per_tahun[tahun] += jumlah

df_total_tahun = pd.DataFrame(
    list(total_per_tahun.items()),
    columns=['tahun', 'total_sampah']
)

df_total_tahun
```

```python
df_tampil = df[['nama_kabupaten_kota', 'tahun', 'jumlah_sampah']]
df_tampil
```

```python
total_kota_tahun = {}

for i, row in df.iterrows():
    key = (row['nama_kabupaten_kota'], row['tahun'])
    jumlah = row['jumlah_sampah']

    if key not in total_kota_tahun:
        total_kota_tahun[key] = 0

    total_kota_tahun[key] += jumlah

df_kota_tahun = pd.DataFrame(
    [(nama_kota, tahun, total)
     for (nama_kota, tahun), total in total_kota_tahun.items()],
    columns=['nama_kabupaten_kota', 'tahun', 'total_sampah']
)

df_kota_tahun
```

```python
df_total_tahun.to_csv("total_sampah_per_tahun.csv", index=False)
df_kota_tahun.to_csv("total_sampah_per_kabupaten_per_tahun.csv", index=False)

with pd.ExcelWriter("rekap_sampah_jawa_barat.xlsx") as writer:
    df_total_tahun.to_excel(writer, sheet_name="Total Per Tahun", index=False)
    df_kota_tahun.to_excel(writer, sheet_name="Per Kabupaten per Tahun", index=False)
```
